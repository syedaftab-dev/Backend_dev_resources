```python
unique_emotions = df['emotion'].unique()
```

```python
emotion_numbers = {}
i=0
for emo in unique_emotions:
  emotion_numbers[emo]=i
  i+=1
```

```python
df['emotion']=df['emotion'].map(emotion_numbers)
```

When `df['emotion'].map(emotion_numbers)` is executed, for each value in the `emotion` column:

1.  The `map()` function looks up the current emotion string (e.g., 'sadness', 'anger') in the `emotion_numbers` dictionary.
2.  It then replaces that emotion string with its corresponding numerical value found in the dictionary (e.g., 'sadness' becomes `0`, 'anger' becomes `1`).

So, the dictionary `emotion_numbers` is the crucial link that defines *which* number replaces *which* word.

## Do cleaning as mention in notebook

```python
# convert to lowercase
df['text']=df['text'].apply(lambda x:x.lower())
```

```python
# removing punctuations like , . ? @ # $ %
import string
def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

### str.maketrans(string,replace string with,delete that string)
### string.punctuation -> contains all punctuations
### here  str.maketrans('','',string.punctuation) - no replacemnt just delete the punctuations

```python
df['text']=df['text'].apply(remove_punc)
```

```python
# removing numbers
def remove_numbers(txt):
  new=""
  for i in txt:
    if not i.isdigit():
      new+=i
  return new
# take a sentence everytime and considering
# word by word ,finding any
# digit in that word then not including that
# word in new framed sentence
```

```python
df['text']=df['text'].apply(remove_numbers)
```

```python
# Remove URls and Links
# same function as above
#and when we get https,// then remove that
```

```python
# Removing Emojis and special symbols


# using ASCII Values - we know ascii
# values of characters and not of emojis
# so ignore them

def remove_emojis(txt):
  new=""
  for i in txt:
    if i.isascii(): # ascii values exist or not
      new+=i
  return new
```

```python
# Removing Stopwords - remove in NLP and keep in Deeplearning accordingly

import nltk #natural language toolkit
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# download these
nltk.download('stopwords') # stopword hai ismain
nltk.download('punkt')# for tokenization
```

```python
stop_words=set(stopwords.words('english'))#getting english stopwords from stopwords library
```

```python
def remove(txt):
    words = txt.split() # or words_tokenize(txt)
    cleaned=[]
    for i in words:
      if i not in stop_words:
        cleaned.append(i)
    return " ".join(cleaned)
```

```python
df['text']=df['text'].apply(remove)
```

## CountVectorizer
### class sklearn.feature_extraction.text.CountVectorizer(*, input='content', encoding='utf-8', decode_error='strict', strip_accents=None, lowercase=True, preprocessor=None, tokenizer=None, stop_words=None, token_pattern='(?u)\\b\\w\\w+\\b', ngram_range=(1, 1), analyzer='word', max_df=1.0, min_df=1, max_features=None, vocabulary=None, binary=False, dtype=<class 'numpy.int64'>)

```python
from sklearn.feature_extraction.text import CountVectorizer

documents=[
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

# n_gram default (1,1) = unigram
vectorizer=CountVectorizer(ngram_range=(3,3))
X=vectorizer.fit_transform(documents)

print("vocabulary",vectorizer.get_feature_names_out())
print("\nBow Matrix:\n",X.toarray())
```

## TF - IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents=[
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

# n_gram default (1,1) = unigram
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(documents)

print("vocabulary",vectorizer.get_feature_names_out())
print("\nBow Matrix:\n",X.toarray())
```
